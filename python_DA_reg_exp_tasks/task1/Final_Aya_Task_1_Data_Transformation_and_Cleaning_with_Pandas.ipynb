## 4.There are also several countries with numbers and/or parenthesis in their name. Be sure to remove these, 


e.g. `'Bolivia (Plurinational State of)'` should be `'Bolivia'`, `'Switzerland17'` should be `'Switzerland'`.

````python
import re
````

````python
def extract(text):
    text =text.split('(')[0]
    return ''.join((i if i.isalpha() or i == ',' else ' ') for i in text).strip()
````

#### Another cleaning function

````python
def cleanCountry(country):
  country=country.split('(')[0]
  result = re.sub(r"[\(\.*?\d?\$'%_)\]]", "", country)
  return result.strip()
````

````python
energy['Country'] = energy['Country'].apply(extract)
  
````
